```python
import pickle
from matplotlib import pyplot as plt
```

```python
history_freeze_first_299_layers_of_inception_v3_and_train_the_rest = pickle.load(open("history_freeze_first_299_layers_of_inception_v3_and_train_the_rest.pkl", "rb"))
# history_freeze_first_299_layers_of_inception_v3_and_train_the_rest
```

```python
def plot_accuracy_vs_epochs(train_acc, val_acc, title):
    plt.title(title)
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.ylim(0.0, 1.1)
    lw = 2
    plt.plot(train_acc, label="Training accuracy", color="darkorange", lw=lw)
    plt.plot(val_acc, label="Validation accuracy", color="navy", lw=lw)
    plt.legend(loc="best")
    plt.show()
```

```python
plot_accuracy_vs_epochs(history_freeze_first_299_layers_of_inception_v3_and_train_the_rest['acc'][:15],
                       history_freeze_first_299_layers_of_inception_v3_and_train_the_rest['val_acc'][:15],
                       "Freeze first 299 layers of inception v3 and train the rest")
```
